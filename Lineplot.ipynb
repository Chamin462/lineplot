# Line plot with multifacets

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
boat=sns.load_dataset("dots")
boat.head()
```

```python
sns.relplot(data=boat,
             x="time",y="firing_rate",
           hue="coherence",size="choice",col="align",kind="line",size_order=["T1","T2"],
           height=5,aspect=.75,facet_kws=dict(sharex=False))
```
